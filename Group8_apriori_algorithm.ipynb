```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('cleaned_vers_data_covid19_vaccine_2020-2022.csv')

# Print the total count of VAERS_ID
total_count = len(df['VAERS_ID'])
print("Total count of VAERS_ID:", total_count)

# Remove duplicate VAERS_ID
df.drop_duplicates(subset='VAERS_ID', inplace=True)

# Print the count after removing duplicates
count_after_duplicates = len(df['VAERS_ID'])
print("Count after removing duplicates:", count_after_duplicates)
```

```python
# Convert the symptom columns to lowercase
df['SYMPTOM1'] = df['SYMPTOM1'].str.lower()
df['SYMPTOM2'] = df['SYMPTOM2'].str.lower()
df['SYMPTOM3'] = df['SYMPTOM3'].str.lower()
df['SYMPTOM4'] = df['SYMPTOM4'].str.lower()

# Concatenate the symptom columns into a single column
symptoms = pd.concat([df['SYMPTOM1'], df['SYMPTOM2'], df['SYMPTOM3'], df['SYMPTOM4']], ignore_index=True)
symptoms[symptoms != 'no symptom']
df['symp'] = symptoms

df['symp'].unique().size

```

```python
# can you find the most common symptom?
df['symp'].value_counts()
# can you find the least common symptom?
df['symp'].value_counts(ascending=True)

top_30_symptoms = df['symp'].value_counts().sum()
print(top_30_symptoms)

less_than_10_symptoms = df['symp'].value_counts()[df['symp'].value_counts() < 10].index.tolist()
print(less_than_10_symptoms)
```
