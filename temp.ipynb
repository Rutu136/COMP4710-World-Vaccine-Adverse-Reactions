```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('cleaned_vers_data_covid19_vaccine_2020-2022_duplicate_removed.csv')

# Extract the specified columns
columns = ['VAERS_ID', 'AGE_YRS', 'SEX', 'DIED', 'L_THREAT', 'HOSPITAL', 'DISABLE', 'OFC_VISIT', 'ER_ED_VISIT', 'OTHER_MEDS', 'CUR_ILL', 'HISTORY', 'ALLERGIES', 'VAX_NAME', 'Symptoms']
df_subset = df[columns]

# Display the extracted columns
df_subset.head()
```

AGE

```python
def map_age_group(age):
    if age < 18:
        return 'children'
    elif age >= 18 and age <= 60:
        return 'adults'
    else:
        return 'seniors'

df_subset['AGE_GROUP'] = df_subset['AGE_YRS'].apply(map_age_group)

df_subset.head()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

df = pd.read_csv('testSubset.csv')#cleaned_vers_data_covid19_vaccine_2020-2022_duplicate_removed.csv')#testSubset.csv')
columns = ['VAERS_ID', 'OTHER_MEDS', 'ALLERGIES', 'OFC_VISIT', 'ER_ED_VISIT', 'HOSPITAL']
df_subset = df[columns].copy()

med_replacement = {'HCTZ': 'BP Meds', 'Lisinopril': 'BP Meds', 'Vitamin C': 'VitC', 'Protonix': 'pantoprazole', 'Atorvastatin': 'cholesterol med', 'Losartan': 'BP Meds', 'hydrochlorothiazide': 'BP Meds', 'none': 'BP Meds', 'No medication': 'BP Meds'}

#convert each list of meds to a string
df_subset['OTHER_MEDS'] = df_subset['OTHER_MEDS'].apply(lambda x: ','.join(x) if isinstance(x, list) else x)

#this creates a new list for each entry by matching the dictionary with the entries 
def process_meds(row):
    meds_list = []
    for med, replacement in med_replacement.items():
        if med in row['OTHER_MEDS']:
            meds_list.append(replacement)
    
    if 'Penicillin' in str(row['ALLERGIES']):
        meds_list.append("Penicillin allergy")

    if 'Y' in str(row['OFC_VISIT']):
        meds_list.append("Saw Doctor")

    if 'Y' in str(row['ER_ED_VISIT']):
        meds_list.append("Went to ER")

    if 'Y' in str(row['HOSPITAL']):
        meds_list.append("Hospitalized")

    return meds_list if meds_list else ['ignore']

df_subset['MED_LIST'] = df_subset.apply(process_meds, axis=1)
df_subset = df_subset.drop('OTHER_MEDS', axis=1)
#print(df_subset.head())

# Step 2: Convert to List of Lists
transactions = df_subset.groupby('VAERS_ID')['MED_LIST'].apply(lambda x: [item for sublist in x for item in sublist]).tolist()
for t in transactions:
    print(t)

# #Step 3: Run Apriori
results = list(apriori(transactions, min_support=0.01, min_confidence=0.01, min_length=2, max_length=2))

# Print the results
for item in results:
    pair = item[0]
    if(len(pair)>1):
        print(item)
        items = [x for x in pair]
        print("Rule: " + items[0] + " -> " + items[1])

```
